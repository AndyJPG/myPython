```python
air_quality_long = pd.read_csv("data/air_quality_long.csv", index_col="date.utc", parse_dates=True)
air_quality_long.head()
```

```python
# long to wide table format
# filter for no2 data only
no2 = air_quality_long[air_quality_long["parameter"] == "no2"]

# use 2 measurements (head) for each location (groupby)
no2_subset = no2.sort_index().groupby(["location"]).head(2)
no2_subset
```

```python
no2_subset.pivot(columns="location", values=["value", "unit"])
```

```python
no2.pivot(columns="location", values="value").plot(figsize=(20, 12))
```

```python
air_quality_long.pivot_table(values="value", index="location", columns="parameter", aggfunc="mean", margins=True)
```

```python
air_quality_long.groupby(["parameter", "location"]).mean()
```

```python
# wide format to long format
no2_pivoted = no2.pivot(columns="location", values="value").reset_index()
no2_pivoted.head()
```

```python
no2.head()
```

```python
no2_pivoted.melt(id_vars="date.utc").head()
```

```python
no2_pivoted.melt(id_vars="date.utc",
                value_vars=["BETR801",
                           "FR04014",
                           "London Westminster"],
                value_name="NO_2",
                var_name="id_location").head()
```
